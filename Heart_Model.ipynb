```python
#Loading all data
import pandas as pd
import numpy as np
data_path='https://s3.amazonaws.com/drivendata/data/54/public/train_values.csv'
train_data=pd.read_csv(data_path)
train_data=pd.DataFrame(train_data)
thel={'normal':1,'fixed_defect':2,'reversible_defect':3}        #categorical encoding of variable for better logloss
train_data.thal = [thel[item] for item in train_data.thal]       
labels=pd.read_csv('https://s3.amazonaws.com/drivendata/data/54/public/train_labels.csv')
labels=pd.DataFrame(labels)
columns=train_data.columns
columns=columns.drop(['patient_id'])
print(columns)
y=labels.heart_disease_present
```

```python
#checking correlation of each data label with the Prediction and deciding features
for x in columns:
    print(train_data[x])
    correlation=(np.corrcoef(train_data[x],y).mean())
    print(correlation)
 #Considering features only with correlation score>0.6(modulus)  
features=['exercise_induced_angina','sex','oldpeak_eq_st_depression','num_major_vessels','chest_pain_type','thal','slope_of_peak_exercise_st_segment']
 
```

```python
#Loading test data and train data
test_data=pd.read_csv('https://s3.amazonaws.com/drivendata/data/54/public/test_values.csv')
test_data=pd.DataFrame(test_data)
thel={'normal':1,'fixed_defect':2,'reversible_defect':3}        
test_data.thal = [thel[item] for item in test_data.thal] 
X=train_data[features]
prediction_X=test_data[features]
```

```python
#Loading Model
from xgboost import XGBRegressor
my_model = XGBRegressor()
my_model.fit(X,y,verbose=False)
prediction = my_model.predict(prediction_X)
```

```python
#Output to csv for submission
output=pd.DataFrame({'patient_id':test_data.patient_id,'heart_disease_present':prediction})
filename = 'SubmissionDeep.csv'
output.to_csv(filename,index=False)
```
